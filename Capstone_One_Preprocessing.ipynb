```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('DW_EDA_Capstone_Project.csv')
df.head()
```

```python
df2 = df.groupby(['InvoiceNo', 'StockCode']).Quantity.sum().unstack().reset_index()
df2 = df2.fillna(0).set_index('InvoiceNo')
df2
```

```python
df2.columns
```

```python
def hot_encode(x):
    if x <= 0:
        return 0
    else:
        return 1
```

```python
df_encoded = df2.applymap(hot_encode)
```

```python
df_two_or_more = df_encoded[(df_encoded > 0).sum(axis=1) >= 2]
```

```python
apriori_results = apriori(df_two_or_more, min_support=.03, use_colnames=True)
apriori_results = apriori_results.sort_values(by='support', ascending=False).reset_index(drop=True)
apriori_results
```

```python
apriori_results['num_items'] = apriori_results['itemsets'].apply(lambda a: len(a))
apriori_results = apriori_results.sort_values(by='num_items', ascending=False)
```

```python
assoc_rule_results = association_rules(apriori_results, metric='lift', min_threshold=1)
assoc_rule_results
```
